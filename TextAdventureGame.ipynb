```python
"""
Created on Wed Feb 1 2023
@author: catarina.jesus
"""
"""
    DocString:
    
    A) Introduction:
    
    This game is based on the movie Back to the Future 1(1985).
    It consists of three stages, as well as, defined functions 
    to start, win and lose.
    The player, Marty Mcfly, travels back to 1955 and needs to
    make some decisions in a way to not change the present.
    
    Round 1: Prevent the father to be or not being hit by the car.
    Round 2: Father meets or not the mother.
    Round 3: The stranger asks questions about the doctor.
    
    The game ends either in success, the presents remain unchanged,
    or in failure, the player´s actions change the present.
    
    B) Known Bugs and/or Errors:
    None.
    
"""

#Import random package to generate random numbers and sys

import random
import sys
import time


start_time=0

#Starting of the game

"""
Asks the player if they want to play with the computer. 
If the player inputs 'yes', then the game continues to the next stage. 
If the player inputs 'no', then the game ends.
"""

def start():
        global start_time
        start_time=time.time()
        #Ask if the player wants to play with the computer.
        user = ""

        #Condition if the player says something different that yes/no.

        while user !="yes" and user!="no" :
            user = input("""
Welcome to Hill Valley, 1985.
You, Marty McFly, met Dr. Emmett Brown, a scientist.
You doctor friend, created a car that was able to travel on the time.
Would you like to travel on the time? (yes/no)
                """).lower()


        #Condition if the player says yes.

            if user == "yes":    
                print("""
\033[1mWelcome to the year 1955!\033[0m""")
                past_time()

        #Condition if the player says no.
            else: 
                print("""
Game Over!""")
                exit(0)

        #First stage when the player decides to play the past time game.

"""
Function for the first stage of the game when the player travels to the past time.
The player makes a decision about an event involving their father and mother. 
Depending on their choice, the game continues to the next stage.
"""

def past_time():

            print("""
Marty, welcome to the past, where you meet your teenage parents.
Your father is crossing the street and he is about to be hit by a car, which is being driven by your mother. 

\033[1mYou need to make a decision regarding this event:

1) Push your father to not be hit by the car and do not meet your mother
2) Let him be hit by the car and meet your mother
\033[0m """)

            choice_1 = input("""
Select 1 or 2: """)

            if "1" in choice_1:
                print("""
You need to find a way for your parents to meet again, otherwise, you will change the present.""") 
                father_meet_mother()

            elif "2" in choice_1:
                print("""
Your father will meet your mother and they will fall in love. 
Now is the time for you to meet the young Doctor to turn back to the present.
                """)
                turn_back_to_present()

            else: 
                print("""
Invalid command. Please select 1 or 2:""")
                past_time()

        #Second stage when the player protect his father of being hit by the car.

"""
Function for the second stage of the game when the player protects their father from being hit by a car.
The player makes a decision about how to make their mother fall in love with their father. 
Depending on their choice, the game continues to the next stage.
"""

def father_meet_mother():

        print("""
Your mother was extremely worried once she hit you.
During that period, she starts being attracted to you.
You need to find a way to change her mind and make her fall in love with your father.

\033[1m You have 2 solutions:

1) Make an arrangement between your parents.
2) Let her fall in love with you
\033[0m""")

        choice_2 = input("""
Select 1 or 2: """)

        if "1" in choice_2:
            print("""
Amazing, your parents fell in love and the present was not changed.\nNow you need to turn back to the present time.""")
            turn_back_to_present()

        elif "2" in choice_2:
            print("""
You lose the game because your parents never fell in love, and you changed the present.""")
            fail()

        else: 
            print("""
Invalid command. Please select 1 or 2: """)
            father_meet_mother()

    #Third stage when the player will turn back to the present time.

"""
Function for the third stage of the game where the player needs to turn back to the present time.
The player is asked a series of random questions by a stranger, and based on their answers, they either pass or lose the game.
"""

def turn_back_to_present():

        print("""
Your task is to find the doctor and help him to build a time machine.
You found some stranger on the street that knows where he is.
Although you need to answer some questions, and depending on your answer he will tell you the trust or not.
        """)


        list_of_questions = ['Did you born before 1955?',
                             'Have you heard about a time machine?',
                             'Which color is the car time machine?']


        asked_questions = []

        while list_of_questions:
            question = random.choice(list_of_questions)
            print(question)
            asked_questions.append(question)
            list_of_questions.remove(question)

            if question == 'Did you born before 1955?':

                while True:
                    choice_3 = input("""
Select yes or no: """).lower()

                    if "no" in choice_3:
                        print("""
Thank you for your honesty!""")
                        break

                    elif "yes" in choice_3:
                        print("""
You lied to me so you lost.""")
                        return fail()

                    else:
                        print("""
Invalid input. Please answer yes or no!""")

            elif question == 'Have you heard about a time machine?':

                while True:
                    choice_4 = input("""
Select yes or no: """).lower()

                    if "yes" in choice_4:
                        print("""
Interesting! You are the time traveler.""")
                        break

                    elif "no" in choice_4:
                        print("""
You lied to me so you lost.""")
                        return fail()

                    else:
                        print("""
Invalid input. Please answer yes or no!""")


            elif question == 'Which color is the car time machine?':

                while True:
                    choice_5 = input("""
Choose between yellow, blue and red.""").lower()

                    if "blue" in choice_5:
                        print("""
You were lucky!""")
                        break

                    elif "red" and "yellow" in choice_5:
                        print("""
You lied to me so you lost.""")
                        return fail()

                    else:
                        print("""
Invalid input. Please answer yellow, blue or red!""")


        print("""You pass the stranger´s test.\nHe told you that the doctor is at the church waiting for you.""")
        final()

"""
Function that is called when the player wins the game.
Prints a congratulations message and exits the game.
"""

def final():
    final_time=time.time()
    global start_time
    if final_time-start_time < 180:
        print("""
\033[1mCongratulations! You won the game, and you were able to turn back to the present without changing it.\033[0m""")
        exit(0)
    else:
        fail()

    #Fourth stage if the player loses.

"""
Function that is called when the player loses the game.
Prints a failure message and exits the game.
"""

def fail():

        print("""
\033[1mYou have failed the game and you would not able to turn back to the present.\033[0m""")
        print("""
\033[1mGame Over!\033[0m""")
        exit(0)

#Starting game!
start()
```
